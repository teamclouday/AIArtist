```python
# select two images
directory = os.path.join("..", "images")
content_file = "clouds-dawn-nature-ocean.jpeg"
style_file = "gray-and-black-marble-slab.jpeg"
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
imshow(load_img(os.path.join(directory, content_file)), os.path.splitext(content_file)[0])
plt.subplot(1, 2, 2)
imshow(load_img(os.path.join(directory, style_file)), os.path.splitext(style_file)[0])
```

```python
model = StyleTransfer(os.path.join(directory, content_file), os.path.join(directory, style_file))
opt = tf.optimizers.Adam(learning_rate=0.02, beta_1=0.99, epsilon=1e-1)
```

```python
model.transfer(opt)
```

```python
model.savefig("test.png")
```
