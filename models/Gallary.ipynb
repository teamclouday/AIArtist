### Art Gallary (Original Images)

```python
import os
import matplotlib.pyplot as plt
from StyleTransfer import imshow, load_img
```

```python
directory = os.path.join("..", "images")
files = []
for filename in os.listdir(directory):
    if not filename.endswith("md"):
        files.append(filename)
nrows = (len(files) // 2) if len(files) % 2 == 0 else (len(files) // 2 + 1)
plt.figure(figsize=(15, 8*nrows))
for i in range(len(files)):
    filename = files[i]
    plt.subplot(nrows, 2, i+1)
    title = os.path.splitext(filename)[0]
    imshow(load_img(os.path.join(directory, filename)), title)
```
