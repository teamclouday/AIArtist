```python
content_image_path = "stretching-white-cat-979247.jpg"
style_image_path = "blue-and-red-abstract-painting-1799912.jpg"
```

```python
plt.subplot(1, 2, 1)
imshow(load_img(content_image_path), title="cat")
plt.subplot(1, 2, 2)
imshow(load_img(style_image_path), title="abstract")
plt.show()
```

```python
model = StyleTransfer(content_image_path, style_image_path)
opt = tf.optimizers.Adam(learning_rate = 0.02,beta_1 = 0.99, epsilon=1e-1)
model.transfer(opt)
model.savefig("normal.png")
```

```python
tf.keras.backend.clear_session()
model = StyleTransfer(content_image_path, style_image_path, content_layers=["block5_conv4"])
opt = tf.optimizers.Adam(learning_rate = 0.02,beta_1 = 0.99, epsilon=1e-1)
model.transfer(opt)
model.savefig("content_layer_block5_conv4.png")
```
